# Exploration Strategies: Entropy vs Random

**Hypo:** Entropy based method create a different behavior for stochastic world.

**Hypo:** Entropy baesd method can't be better in extreme caese. TODO specify exterme.

We'll investigate these questions in the following minimal scenario (similar to a bandit setting):

 - The world consists of doors.
 - A door can be opened with probability $p$.
 - In every turn the agent picks a door and tries to open it.

```python
class Door(object):
    def __init__(self, prob_open, name, verbose=False):
        self.p = prob_open
        self.name = name
        self.verbose = verbose
        
    def push(self):
        result = random.random() < self.p
        if self.verbose:
            print("{} (prob_open={}) {}".format(
                self.name, self.p,
                "opened" if result else "didn't open"
            ))
        return result
    
doors = [Door(prob_open=1, name="door 1"), Door(prob_open=.5, name="door 2")]
```

```python
d1 = doors[1]
print("opened?:", d1.push(), d1.push(), d1.push())
```

The agent has a belief over the doors.

```python
class DoorBel(object):
    """"DoorBel is a beta dist."""
    def __init__(self, opened=1, closed=1):
        # uninformed prior
        self.opened = opened
        self.closed = closed
        
    def __str__(self):
        return "Beta(opened={}, closed={}) --> H={}".format(
            self.opened, self.closed, ss.beta(self.opened, self.closed).entropy()
        )
    
    def entropy(self):
        return ss.beta(self.opened, self.closed).entropy()
    
    def update(self, opened):
        if opened:
            self.opened += 1
        else:
            self.closed += 1

    def mean(self):
        return ss.beta.mean(self.opened, self.closed)
    
    def likelihood(self, loc):
        return ss.beta.pdf(loc, self.opened, self.closed)

    
world_model = [DoorBel() for i in range(2)]
```

Let't push the doors a couble of times just to see if it works. The entropy of the deterministic door always must be decreasing (in a deterministic manner, i.e. two runs are always the same). This is not the case for the stochastic door: it can increase and runs should differ.

```python
world_model = [DoorBel() for i in range(2)]

runs = 50
entropy = np.empty((2, runs))

for run in range(runs):
    for idx, door in enumerate(doors):

        opened = door.push()
        world_model[idx].update(opened)
    
        ent = world_model[idx].entropy()
        entropy[idx, run] = ent
        # print(world_model[idx])
    
fig, ax = plt.subplots()
ax.plot(entropy[0, :], label="door 1");
ax.plot(entropy[1, :], label="door 2");
ax.legend();
```

# Random Strategy
The random exploration strategy selects doors randomly.

```python
def run(select_strategy, world_model=None):
    if world_model is None:
        world_model = [DoorBel() for i in range(2)]

    runs = 50
    history = []
    dom = np.linspace(0, 1)
    entropy = np.empty((2, runs))
    err = np.empty((2, runs))
    score = np.empty((2, runs))
    Z_door1 = np.empty((runs, runs))
    Z_door2 = np.empty((runs, runs))


    for run in range(runs):
        idx = select_strategy(world_model)
        door = doors[idx]
        opened = door.push()
    
        world_model[idx].update(opened)
        
        # calculate statistics
        for i, bel in enumerate(world_model):
            entropy[i, run] = bel.entropy()
            err[i, run] = abs(doors[i].p - ss.beta.mean(bel.opened, bel.closed))
            score[i, run] = ss.beta.pdf(doors[i].p, bel.opened, bel.closed)
        Z_door1[:, run] = ss.beta.pdf(dom, world_model[0].opened, world_model[0].closed)
        Z_door2[:, run] = ss.beta.pdf(dom, world_model[1].opened, world_model[1].closed)

        history.append(world_model)
        
    return world_model, entropy, Z_door1, Z_door2, history, err, score
```

```python
def strategy_random(world_model):
    return random.randint(0, 1)
```

```python
def analyse(entropy, err, score, doors, world_model):
    fig, (ax1, ax2) = plt.subplots(2, figsize=(12, 8))
    
    ax1.plot(entropy[0, :], label="door 1")
    ax1.plot(entropy[1, :], label="door 2")
    ax1.set_title("Random Strategy: Entropy over time")
    ax1.grid()
    ax1.legend();
    
    ax2.plot(err[0, :])
    ax2.plot(err[1, :])
    ax2.set_title("Random Strategy: Error over time")
    ax2.grid()
    
    print("Sum error:", err.sum())
    for i, (door, bel) in enumerate(zip(doors, world_model)):
        print("door {}: real: {} est: {} err: {} likelihood: {}".format(
            i, door.p, bel.mean(), abs(door.p - bel.mean()), bel.likelihood(door.p)))
```

```python
def plot_pdf3d(dom, runs, Z_door1, Z_door2):
    
    #fig, ax = plt.subplots()
    #ax.set_title("final distributions")
    #for door_bel in world_model:
    #    ax.plot(dom, ss.beta.pdf(dom, door_bel.opened, door_bel.closed))
        
    X = np.column_stack([dom for _ in range(runs)])
    Y = np.column_stack([np.ones((runs, 1)) * i for i in range(runs)])

    fig = plt.figure(figsize=(18, 18))

    ax = fig.add_subplot(221, projection='3d')
    ax.plot_wireframe(X, Y, Z_door1, color="blue", alpha=.2)
    ax = fig.add_subplot(223)
    for i in range(0, 50, 5):
        ax.plot(dom, Z_door1[:, i], color="blue", alpha=i/50)

    ax = fig.add_subplot(222, projection='3d')
    ax.plot_wireframe(X, Y, Z_door2, color="green", alpha=.2)
    ax = fig.add_subplot(224)
    for i in range(0, 50, 5):
        ax.plot(dom, Z_door2[:, i], color="green", alpha=i/50)
```

```python
world_model, entropy, Z1, Z2, history, err, score = run(strategy_random)
analyse(entropy, err, score, doors, world_model)
plot_pdf3d(dom, runs, Z1, Z2)
```

As one expects the deterministic door is less interesting. The entropy can be decreased faster.

# Entropy-based Exploration
Now, let's see if the entropy-based strategy behaves diffrently:

```python
def strategy_entropy(world_model):
    # select max entropy door
    max_ent = -999999
    for i, door in enumerate(world_model):
        ent = door.entropy()
        if ent > max_ent:
            max_ent = ent
            idx = i
    return idx
```

```python
world_model, entropy, Z1, Z2, history, err, score = run(strategy_entropy)
analyse(entropy, err, score, doors, world_model)
plot_pdf3d(dom, runs, Z1, Z2)
```

Maybe we should not look at the maximum entropy but the maximum expected entropy change.

```python
def strategy_expected_change_of_entropy(world_model):

    max_diff = -999999
    for i, bel in enumerate(world_model):
        H = bel.entropy()
        # prob for outcome a
        P_a = bel.mean()
        # entropy for outcome a
        H_a = ss.beta.entropy(bel.opened + 1, bel.closed)
        # prob for outcome b
        P_b = 1 - P_a
        # entropy for outcome b
        H_b = ss.beta.entropy(bel.opened, bel.closed + 1)
        # new estimated entropy
        H_est = (P_a * H_a + P_b * H_b)
        # change of H
        diff = H - H_est
        if False:
            print("=================================")
            print("H:    ", H)
            print("P_a:  ", P_a)
            print("H_a:  ", H_a)
            print("P_b:  ", P_b)
            print("H_b:  ", H_b)
            print("H_est:", H_est)
            print("diff: ", diff)

        if diff > max_diff:
            max_diff = diff
            idx = i
    return idx

world_model = [DoorBel(5, 5), DoorBel(7, 1)]
idx = strategy_expected_change_of_entropy(world_model)
print("picked", idx)
```

```python
world_model, entropy, Z1, Z2, history, err, score = run(strategy_expected_change_of_entropy)
analyse(entropy, err, score, doors, world_model)
plot_pdf3d(dom, runs, Z1, Z2)
```

# Average over multiple runs

```python
def multi_run(strategy, name):
    print(name)
    
    n = 50
    total_err = np.zeros((n, 50))
    total_err1 = np.zeros((n, 50))
    total_err2 = np.zeros((n, 50))
    for i in range(n):
        world_model, entropy, Z1, Z2, history, err, score = run(strategy_entropy)
        total_err[i, :] = err[0, :] + err[1, :]
        total_err1[i, :] = err[0, :]
        total_err2[i, :] = err[1, :]
        
    fig, ax = plt.subplots(figsize=(12, 12))
    
    mu1 = total_err1.mean(axis=0)
    sigma1 = total_err1.std(axis=0)
    ax.plot(mu1)
    ax.fill_between(range(50), mu1+sigma1, mu1-sigma1, alpha=.2, color="blue")

    mu2 = total_err2.mean(axis=0)
    sigma2 = total_err2.std(axis=0)
    ax.plot(mu2)
    ax.fill_between(range(50), mu2+sigma2, mu2-sigma2, alpha=.2, color="green")
    
    mu = total_err.mean(axis=0)
    ax.plot(mu, label="total")


    ax.set_title("Error over time for Strategy " + name);

    print("MSE door 0:", np.mean(mu1**2))
    print("MSE door 1:", np.mean(mu2**2))
    print("MSE total :", np.mean(mu**2))
    return mu
```

```python
err_random = multi_run(strategy_random, "Random");
```

```python
err_ent = multi_run(strategy_entropy, "Entropy");
```

```python
err_exp_change = multi_run(strategy_expected_change_of_entropy, "Expected Entropy Change");
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(err_random, label="random")
ax.plot(err_ent, label="entropy")
ax.plot(err_exp_change, label="exp. change")
ax.set_title("Error over time for different exploration strategies")
ax.grid()
ax.legend();
```
